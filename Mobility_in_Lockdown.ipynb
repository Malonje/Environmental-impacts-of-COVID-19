```python

import pandas as pd
import io
```

```python
import requests
url = 'https://www.gstatic.com/covid19/mobility/Global_Mobility_Report.csv'


gfile = requests.get(url, allow_redirects=True)
open('gmobility_data.csv', 'wb').write(gfile.content)
```

```python
import csv
with open('gmobility_data.csv') as csvfile:
    reader = csv.DictReader(csvfile, delimiter=",", quotechar='"')
    print(reader.fieldnames)
```

```python
df=pd.read_csv('/content/gmobility_data.csv')
```

```python
data=df
```

```python
data = data[ (data['country_region_code'] == 'IN') & (data['sub_region_1']=='Delhi') & (data['sub_region_2']=='Central Delhi') ]
```

```python
data=data[:321]
```

```python
y=data['retail_and_recreation_percent_change_from_baseline'].tolist()
x=data['date'].tolist()
```

```python
from matplotlib import pyplot as plt
plt.plot(x,y)
#specify x-axis locations
x_ticks = [0, 30, 60, 90, 120, 150, 180, 210, 240, 270, 300, 330, 360]
x_labels = ['feb','march','april','may','june','july','august','september','october','november','december']
plt.xticks(ticks=x_ticks, labels=x_labels)

plt.ylabel("% CHANGE IN MOBILITY")
plt.title('MOBILITY IN CENTRAL DELHI REGION 2020')
plt.show()
```
